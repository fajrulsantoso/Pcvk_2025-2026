 **3**  Buatlah fungsi dilasi serta tampilkan hasil dilasi pada citra dengan Structuring Element
Dilasi (SED) berbentuk cross 3 x 3 tanpa menggunakan library morphology dari
openCV, sehingga menghasilkan tampilan seperti di bawah ini

```python

# Fungsi Dilasi Tanpa Library Khusus
def dilasi_citra(F, w):
    # Ukuran citra
    p, q = F.shape
    imgD = np.zeros((p, q), dtype=np.uint8)

    # Elemen struktur dilasi (SED)
    SED = np.ones((w, w), dtype=np.uint8)
    constant1 = (w - 1) // 2

    # Proses Dilasi
    for i in range(constant1, p - constant1):
        for j in range(constant1, q - constant1):
            temp = F[i - constant1 : i + constant1 + 1, j - constant1 : j + constant1 + 1]
            product = temp * SED
            imgD[i, j] = np.max(product)

    return imgD

# Membaca citra grayscale
img_d = cv2.imread('/content/drive/MyDrive/Pcvk/plat_nomer.jpg', 0)

# Threshold untuk binerisasi citra
ret, thresh = cv2.threshold(img_d, 125, 200, cv2.THRESH_BINARY)

# Ukuran elemen struktur
w = 5

# Proses dilasi
imgD = dilasi_citra(thresh, w)

# Tampilkan hasil
plt.figure(figsize=(10,5))
plt.subplot(121), plt.imshow(img_d, cmap='gray')
plt.title('Citra Awal'), plt.axis('off')

plt.subplot(122), plt.imshow(imgD, cmap='gray')
plt.title('Citra Hasil Dilasi'), plt.axis('off')

plt.show()
```

**5**. Buatlah fungsi erosi serta tampilkan hasil erosi pada citra dengan Structuring Element
Erosi (SE) berbentuk square 5 x 5 tanpa menggunakan library morphology dari
openCV, sehingga menghasilkan tampilan seperti di bawah ini:

```python
def erosi_citra(F, k):
    #size image
    m, n = F.shape
    #Generate structure element erosi (SE)
    SE = np.ones((k, k), dtype=np.uint8)
    constant = (k - 1) // 2
    imgE = np.zeros((m, n), dtype=np.uint8)
    #Proses Erosi
    for i in range(constant, m - constant):
        for j in range(constant, n - constant):
            temp = F[i - constant : i + constant + 1, j - constant : j + constant + 1]
            product = temp * SE
            imgE[i, j] = np.min(product)
    return imgE


img_e = cv2.imread('/content/drive/MyDrive/Pcvk/plat_nomer.jpg', 0)
ret1, thresh1 = cv2.threshold(img_e, 175, 225, cv2.THRESH_BINARY)
k = 5
imgE = erosi_citra(thresh1, k)

plt.subplot(131), plt.imshow(img_e, cmap='gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(imgE, cmap='gray')
plt.title('Citra Hasil Erosi'), plt.xticks([]), plt.yticks([])

plt.show()
```

**9.** Buatlah program implementasi operasi dilasi dengan Structuring Element berbentuk

**10.** Buatlah program implementasi operasi dilasi dengan Structuring Element berbentuk
Circular 3 x 3 dan 5 x 5, sehingga menghasilkan tampilan seperti di bawah ini:

```python
# ======================================================
# DILASI CIRCULAR SED
# ======================================================
def dilasi_citra(F, w):
    # size image
    p, q = F.shape
    imgD = np.zeros((p, q), dtype=np.uint8)

    # Generate structure element dilasi (SED) berbentuk lingkaran
    SED = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (w, w))
    constant1 = (w - 1) // 2

    # Proses Dilasi
    for i in range(constant1, p - constant1):
        for j in range(constant1, q - constant1):
            temp = F[i - constant1 : i + constant1 + 1, j - constant1 : j + constant1 + 1]
            product = temp * SED
            imgD[i, j] = np.max(product)

    return imgD


# ======================================================
# Circle 3
# ======================================================
print('SED = Circle 3')
img_d = cv2.imread('/content/drive/MyDrive/Pcvk/plat_nomer.jpg', 0)
ret, thresh = cv2.threshold(img_d, 127, 225, cv2.THRESH_BINARY)
w = 3
imgD = dilasi_citra(thresh, w)

plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(img_d, cmap='gray')
plt.title('Citra Awal (Circle 3)'), plt.axis('off')
plt.subplot(122), plt.imshow(imgD, cmap='gray')
plt.title('Citra Hasil Dilasi (Circle 3)'), plt.axis('off')
plt.show()


# ======================================================
# Circle 5
# ======================================================
print('SED = Circle 5')
img_d = cv2.imread('/content/drive/MyDrive/Pcvk/plat_nomer.jpg', 0)
ret, thresh = cv2.threshold(img_d, 127, 225, cv2.THRESH_BINARY)
w = 5
imgD = dilasi_citra(thresh, w)

plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(img_d, cmap='gray')
plt.title('Citra Awal (Circle 5)'), plt.axis('off')
plt.subplot(122), plt.imshow(imgD, cmap='gray')
plt.title('Citra Hasil Dilasi (Circle 5)'), plt.axis('off')
plt.show()
```

**11.** Buatlah program implementasi operasi erosi dengan Structuring Element berbentuk
Rectangle 3 x 5 dan 5 x 7, sehingga menghasilkan tampilan seperti di bawah ini

```python
# ======================================================
# EROSI RECTANGLE SE
# ======================================================
def erosi_citra(F, p, l):
    # size image
    m, n = F.shape

    # Generate structure element erosi (SE) berbentuk persegi panjang
    SE = cv2.getStructuringElement(cv2.MORPH_RECT, (p, l))  # SE persegi panjang
    constant1 = (l - 1) // 2
    constant2 = (p - 1) // 2

    imgE = np.zeros((m, n), dtype=np.uint8)

    # Proses Erosi
    for i in range(constant1, m - constant1):
        for j in range(constant2, n - constant2):
            temp = F[i - constant1 : i + constant1 + 1, j - constant2 : j + constant2 + 1]
            product = temp * SE
            imgE[i, j] = np.min(product)

    return imgE


# ======================================================
# SE Rectangle 3x5
# ======================================================
print('SE = Rectangle 3x5')
img_e = cv2.imread('/content/drive/MyDrive/Pcvk/plat_nomer.jpg', 0)
ret1, thresh1 = cv2.threshold(img_e, 127, 225, cv2.THRESH_BINARY)
p = 5
l = 3
imgE = erosi_citra(thresh1, p, l)

plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(img_e, cmap='gray')
plt.title('Citra Awal (3x5)'), plt.axis('off')
plt.subplot(122), plt.imshow(imgE, cmap='gray')
plt.title('Citra Hasil Erosi (3x5)'), plt.axis('off')
plt.show()


# ======================================================
# SE Rectangle 5x7
# ======================================================
print('SE = Rectangle 5x7')
img_e = cv2.imread('/content/drive/MyDrive/Pcvk/plat_nomer.jpg', 0)
ret1, thresh1 = cv2.threshold(img_e, 127, 225, cv2.THRESH_BINARY)
p = 7
l = 5
imgE = erosi_citra(thresh1, p, l)

plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(img_e, cmap='gray')
plt.title('Citra Awal (5x7)'), plt.axis('off')
plt.subplot(122), plt.imshow(imgE, cmap='gray')
plt.title('Citra Hasil Erosi (5x7)'), plt.axis('off')
plt.show()
```

**12.** Buatlah program implementasi operasi erosi dengan Structuring Element berbentuk
Line Vertikal 3 dan 5, sehingga menghasilkan tampilan seperti di bawah ini: (Gunakan
gambar j.png)

```python
# ======================================================
# EROSI LINE VERTIKAL SE
# ======================================================
def erosi_citra(F, k):
    # size image
    m, n = F.shape

    # Generate structure element erosi (SE) line vertikal
    SE = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (1, k))  # SE line vertikal
    constant = (k - 1) // 2
    imgE = np.zeros((m, n), dtype=np.uint8)

    # Proses Erosi
    for i in range(constant, m - constant):
        for j in range(constant, n - constant):
            temp = F[i - constant : i + constant + 1, j - constant : j + constant + 1]
            product = temp * SE
            imgE[i, j] = np.min(product)

    return imgE


# ======================================================
# SE Line Vertikal 3
# ======================================================
print('SE = Line Vertikal 3')
img_e = cv2.imread('/content/drive/MyDrive/Pcvk/plat_nomer.jpg', 0)
ret1, thresh1 = cv2.threshold(img_e, 127, 225, cv2.THRESH_BINARY)
k = 3
imgE = erosi_citra(thresh1, k)

plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(img_e, cmap='gray')
plt.title('Citra Awal (Line Vertikal 3)'), plt.axis('off')
plt.subplot(122), plt.imshow(imgE, cmap='gray')
plt.title('Citra Hasil Erosi (Line Vertikal 3)'), plt.axis('off')
plt.show()


# ======================================================
# SE Line Vertikal 5
# ======================================================
print('SE = Line Vertikal 5')
img_e = cv2.imread('/content/drive/MyDrive/Pcvk/plat_nomer.jpg', 0)
ret1, thresh1 = cv2.threshold(img_e, 127, 225, cv2.THRESH_BINARY)
k = 5
imgE = erosi_citra(thresh1, k)

plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(img_e, cmap='gray')
plt.title('Citra Awal (Line Vertikal 5)'), plt.axis('off')
plt.subplot(122), plt.imshow(imgE, cmap='gray')
plt.title('Citra Hasil Erosi (Line Vertikal 5)'), plt.axis('off')
plt.show()
```

**15.** Skeleton

```python
import cv2 as cv
from matplotlib import pyplot as plt
from skimage import filters
from skimage.morphology import skeletonize
# read image
img = cv.imread('/content/drive/MyDrive/Pcvk/peppers.jpg', 0)

binary = img > filters.threshold_triangle(img)
# true false to one
binary_cp = binary.copy()
binary_cp[binary_cp == True] = 1
binary_cp[binary_cp == False] = 0
# skeletonize image
skeleton = skeletonize(binary_cp)
# print images
plt.figure(figsize=(20,20))
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(skeleton,cmap = 'gray')
plt.title('skeleton Image'), plt.xticks([]), plt.yticks([])
plt.savefig("binary and bin_skeleton.png")
plt.show()
```

**16.** Skeleton Inverse

```python
from skimage import io, color, filters
from skimage.morphology import skeletonize
from matplotlib import pyplot as plt

# ======================================================
# SKELETONIZATION (PENULANGAN CITRA)
# ======================================================

# Membaca citra dan ubah ke grayscale
img_rgb = io.imread('/content/drive/MyDrive/Pcvk/peppers.jpg')
img = color.rgb2gray(img_rgb)

# Mengubah citra menjadi biner (area gelap menjadi True)
binary = img < filters.threshold_triangle(img)

# Melakukan skeletonisasi
skeleton = skeletonize(binary)

# Menampilkan hasil
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(img, cmap='gray')
ax[0].set_title('Citra Grayscale')
ax[0].axis('off')

# Skeleton dengan colormap terbalik agar lebih kontras
ax[1].imshow(skeleton, cmap='gray_r')
ax[1].set_title('Citra Skeleton (Inverse)')
ax[1].axis('off')

plt.tight_layout()
plt.show()
```
